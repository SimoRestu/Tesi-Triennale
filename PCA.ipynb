```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from numpy.linalg import inv
```

Define covariance matrix

```python
alpha = 0.8
cov = np.array([[1,alpha],[alpha,1]])
```

```python
cov_inv = inv(cov)
```

Sample around zero with the above covariance matrix

```python
mean = np.array([0,0])
```

```python
x1, x2 = np.random.multivariate_normal(mean,cov,10_000).T
X = np.array([x1, x2]).T
```

```python
plt.scatter(x1,x2)
plt.show()
```

Use SVD to create new variables

```python
U, Sigma, Vt = np.linalg.svd(cov_inv)
```

```python
Y = U @ X.T
```

```python
y1, y2 = Y
```

```python
Y_std = 1/Sigma**0.5
```

```python
plt.scatter(y1, y2)
plt.show()
```
